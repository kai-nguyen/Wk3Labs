```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

Sachouse = pd.read_csv('Sacramentorealestatetransactions.csv')

Sachouse.head()
```

```python
Sachouse.describe()
```

```python
# Use pandas to look through the data. Plot the variables as histograms or pairs in scatter plots as 
# needed with matplotlib, pandas, or seaborn until you understand each one.

plt.hist(Sachouse.price, bins=15, color='g')
plt.xlabel('Price')
plt.ylabel('Distribution')
plt.show()
```

```python
plt.hist(Sachouse.beds, bins=8, color='r')
plt.xlabel('Number of bedrooms')
plt.show()
```

```python
plt.hist(Sachouse.sq__ft, bins=15)
plt.xlabel('Square footage of homes')
plt.show()
```

```python
plt.scatter(Sachouse.sq__ft, Sachouse.price, marker='+', color='g')
plt.xlabel('Square footage of Sacramento houses')
plt.ylabel('Price of houses')
plt.show()
```

```python
plt.scatter(Sachouse.zip, Sachouse.price, marker='*', color='c')
plt.xlabel('Zip Code')
plt.ylabel('Price of houses')
plt.show()
```

```python
# Visualize the Data: The data set contains a number of variables that may be correlated with the price of the 
# properties. Make plots of the relevant variables versus the column "price". You can use pandas, matplotlib, or 
# seaborn.

import seaborn as sns
from matplotlib import pyplot as plt

x = Sachouse['beds']
y = Sachouse['price']

plt.scatter(x, y, marker='*', color='m')
plt.xlabel("Number of Bedrooms")
plt.ylabel("Property Price")
plt.show()
```

```python
sns.pairplot(Sachouse, x_vars=['beds','sq__ft','zip'], y_vars=['price'], size=5)
```
