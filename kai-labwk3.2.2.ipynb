#### Evaluating Model Fit
##### Let's look at some data. Scikit-learn can compute r2 for us, so let's explore some actual data.

```python
# Let's load in some packages

% matplotlib inline

from matplotlib import pyplot as plt
# Make the plots bigger
plt.rcParams['figure.figsize'] = 10, 10

import numpy as np
import pandas as pd
from scipy import stats
from sklearn import datasets, linear_model

# Load the Boston Housing dataset
boston = datasets.load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)

# Take a look at the data again
df.head()
```

```python
print boston.DESCR
```

### First example
##### Let's pick two variables and plot them against each other with a best fit line. For example, let's see if the Pupil-Teacher ratio by town and the age of a property are related.

```python
# Fit a line

lm = linear_model.LinearRegression()

X = df[["PTRATIO"]]
y = df["AGE"]

model = lm.fit(X, y)
predictions = lm.predict(X)

# Plot the data and the best fit line
## The data
plt.scatter(X, y)
## The line / model
plt.plot(X, predictions)
plt.ylabel("AGE")
plt.xlabel("PTRATIO")
```

#### The data doesn't appear to be all that linear so we should find that the model fit is poor. Let's calculate the r2 value.

```python
# Scikit-learn makes it easy to get the score
print "r^2:", model.score(X,y)
```

#### In this case the r2 value is close to zero, as expected.

```python

# Here's some quadratic data with a bit of noise
import scipy

noise = scipy.stats.norm(0, 4)
data = [(x, 3*x*x - 5*x + 3 + noise.rvs()) for x in np.arange(-6, 2, 0.5)]

xs = [x for (x, y) in data]
ys = [y for (x, y) in data]
```

```python
# Now let's fit a quadratic model with numpy
# polyfit gives us the coefficients of the best fit polynomial
coef = np.polyfit(xs, ys, deg=2)
# polyvals gives us the polynomial function for these coefficients
predictions = np.polyval(coef, xs)

# Let's plot the model
plt.scatter(xs, ys)
plt.plot(xs, predictions)

# Here's another way to get r^2 from scipy
from sklearn import metrics
metrics.r2_score(ys, predictions)
```

```python
# Now let's fit a quadratic model with numpy
# polyfit gives us the coefficients of the best fit polynomial
coef = np.polyfit(xs, ys, deg=1)

predictions = np.polyval(coef, xs)

plt.scatter(xs, ys)
plt.plot(xs, predictions)

metrics.r2_score(ys, predictions)
```

#### Exercise
##### Now let's look at the effect of outliers. Just one outlier can really skew your models. Let's add outlier to our data and redo both the quadratic and the linear fits. Which model type drops the most in r2?

```python
xs.append(2)
ys.append(120)
```

```python
# Repeat the Quadratic fit

coef = np.polyfit(xs, ys, deg=2)

predictions = np.polyval(coef, xs)

plt.scatter(xs, ys)
plt.plot(xs, predictions)

metrics.r2_score(ys, predictions)
```

```python
# And the linear fit

coef = np.polyfit(xs, ys, deg=1)

predictions = np.polyval(coef, xs)

plt.scatter(xs, ys)
plt.plot(xs, predictions)

metrics.r2_score(ys, predictions)
```
