```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Download the data, save to a file called "housing.data"
import urllib
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
urllib.urlretrieve (data_url, "housing.data")
```

```python
names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
         "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
```

```python
import pandas as pd

data = pd.read_csv("housing.data", header=None, names=names, delim_whitespace=True)

# Take a look at the first few rows
data.head()
```

```python
# Compute the number of rows
data.count()
```

```python
import seaborn as sns

# Plot using pandas
data.plot.scatter(x='CRIM', y='MEDV')
```

#### How do we select a column of data from a pandas DataFrame?

##### We can select a column of data from a pandas DataFrame by specifying the parameter for the methods pd.DataFrame() or in data.plot.scatter(x='name of column', y='name of column'). Depending on the methods we employ, certain parameters may or may not work, but in this case, selecting the column is easily done by specifying the names of the column.

```python
# Knowledge Check Solution

from matplotlib import pyplot as plt

# Plot using matplotlib
plt.scatter(data["CRIM"], data["MEDV"], color='g')
plt.xlabel("Crime per capita")
plt.ylabel("Median house value in $1000s")
plt.show()
```

#### Question: Which variables seem to correlate well?

##### The variables that seem to have a good correlation with median house value are: Avg number of rooms (positive linear correlation), percentage of lower status of population (negative linear correlation), and crime. Another thing to note is how many different factors play into the value of houses in Boston. The age of a house does play a role in its value, i.e., the older the houses are the lower in value they tend to have outside of some outliers.
